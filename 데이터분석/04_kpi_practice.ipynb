```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings

warnings.filterwarnings('ignore', category=UserWarning)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
customer_satisfaction = pd.read_csv('./customer_satisfaction.csv')
customer_transactions = pd.read_csv('./customer_transactions.csv')
marketing_performance = pd.read_csv('./marketing_performance.csv')
```

```python
customer_satisfaction['contact_date'] = pd.to_datetime(customer_satisfaction['contact_date'])
customer_transactions['registration_date'] = pd.to_datetime(customer_transactions['registration_date'])
customer_transactions['transaction_date'] = pd.to_datetime(customer_transactions['transaction_date'])
marketing_performance['month'] = pd.to_datetime(marketing_performance['month'])
```

2. 채널별 CAC 계산
    - 2024년 데이터만 사용 - (2024년 마케팅데이터만 사용 -> 24년 채녈별 마케팅비용/24년 신규 획득 고객 수)
    - 각 채널별 총 마케팅 비용 집계
    - 각 채널별 신규 고객 수 집계 (2024년 첫 거래 기준)
    - CAC = 총 마케팅 비용 / 신규 고객 수
    - organic과 referral 채널은 CAC = 0으로 처리

```python
cac_per_channel = marketing_performance[marketing_performance['month'].dt.year == 2024].groupby('channel').agg({
    'marketing_spend': 'sum',
    'new_customers': 'sum'
}).rename(columns={
    'marketing_spend': 'marketing_spend_sum',
    'new_customers': 'new_customers_sum'
}).reset_index()

cac_per_channel['cac_2024'] = cac_per_channel.apply(
    lambda row: 0 if row['channel'] in ['organic', 'referral']
    or row['new_customers_sum'] == 0
    else round(row['marketing_spend_sum'] / row['new_customers_sum'], 2),
    axis=1
)

cac_per_channel = cac_per_channel.sort_values('cac_2024', ascending=False).reset_index(drop=True)
cac_per_channel
```

3. 고객별 LTV 계산 - 전체 구매 이력 기반

```python
cust_ltv = customer_transactions.groupby('customer_id').agg({
    'order_value': 'sum',
    'transaction_date': ['count', 'max'],
    'registration_date': 'first'
})

cust_ltv.columns = ['_'.join(col) for col in cust_ltv.columns]
cust_ltv = cust_ltv.rename(columns={
    'order_value_sum': 'total_revenue',
    'transaction_date_count': 'transaction_count',
    'registration_date_first': 'registration_date'
}).reset_index()

cust_ltv['cust_lifespan_d'] = (cust_ltv['transaction_date_max'] - cust_ltv['registration_date']).dt.days + 1

# 예외값 제거(Boolean Masking)
cust_ltv = cust_ltv[(cust_ltv['cust_lifespan_d'] > 0) & (cust_ltv['transaction_count'] > 0)]

cust_ltv['total_ltv'] = cust_ltv['total_revenue']
cust_ltv['ltv_per_trans'] = (cust_ltv['total_revenue'] / cust_ltv['transaction_count']).round(2)
cust_ltv['ltv_per_days'] = (cust_ltv['total_revenue'] / cust_ltv['cust_lifespan_d']).round(2)
cust_ltv
```

4. 채널별 LTV 및 ROI 분석 - (`24년 신규 고객들의 평균 LTV / 24년 CAC`)    
    - 채널별 평균 LTV 계산
    - ROI = LTV / CAC 계산 (organic, referral은 무한대 처리)
    - payback period = CAC / (월평균 구매금액) 계산

```python
revenue_per_channel = customer_transactions.groupby([customer_transactions['transaction_date'].dt.to_period('M').dt.to_timestamp(), 'acquisition_channel']).agg({
    'customer_id': 'count',
    'order_value': 'sum'
}).rename(columns={
    'customer_id': 'order_count'
}).reset_index()

channel_summary = pd.merge(marketing_performance, revenue_per_channel, how='left', left_on=['month', 'channel'], right_on=['transaction_date', 'acquisition_channel'])
channel_summary = channel_summary.drop(['transaction_date', 'acquisition_channel'], axis=1)
channel_summary
```

    - 채널별 ROI 막대 차트
    - 마케팅 예산 재배분 제안 (현재 vs 최적 배분)
